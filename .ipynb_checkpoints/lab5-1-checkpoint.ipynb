```python
IMG_SIZE = (256, 256)
seed = 3

def set_seed(seed=42):
    os.environ['PYTHONHASHSEED'] = str(seed)
    tf.keras.utils.set_random_seed(seed)
    tf.config.experimental.enable_op_determinism()

set_seed(seed)

full_ds = tf.keras.utils.image_dataset_from_directory('data#1/all', image_size = IMG_SIZE, seed=seed)
```

```python
data_iterator = full_ds.as_numpy_iterator()
current_batch = data_iterator.next()

def show_some_picture_in_batch(batch, number_of_picture, max_cols=5):
    cols = min(number_of_picture, max_cols)
    rows = math.ceil(number_of_picture / cols)
    fig, axes = plt.subplots(nrows=rows, ncols=cols, figsize=(2*cols, 2*rows))
    
    if rows == 1:
        axes = [axes]
    if cols == 1:
        axes = [[ax] for ax in axes]
    axes = [ax for row in axes for ax in row]  

    for idx in range(number_of_picture):
        img = batch[0][idx]
        label = batch[1][idx]
        axes[idx].imshow(img.astype(int))
        axes[idx].title.set_text(str(label))
        axes[idx].axis('off')

    for idx in range(number_of_picture, len(axes)):
        axes[idx].axis('off')

    plt.tight_layout()
    plt.show()
    
show_some_picture_in_batch(current_batch, 20)
```

<center>Попередня обробка даних</center>

```python
full_ds = full_ds.map(lambda x, y: (x / 255.0, y))
```

```python
total_count = tf.data.experimental.cardinality(full_ds).numpy()
print("Total count of batches", total_count)

train_size = 0.8 * total_count
val_test_size = 0.1 * total_count
```

```python
train_ds = full_ds.take(train_size)
val_ds = full_ds.skip(train_size).take(val_test_size)
test_ds = full_ds.skip(train_size + val_test_size)

print("Train  size:", len(train_ds))
print("Test  size:", len(test_ds))
print("Valid size:", len(val_ds))
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Dense, Flatten, Dropout
from tensorflow.keras import layers, models
from tensorflow.keras.callbacks import EarlyStopping
import numpy as np
```

```python
model = Sequential()

model.add(layers.Input(shape=(IMG_SIZE[0], IMG_SIZE[1], 3))) 

model.add(Conv2D(16, (1,1), 1, activation='relu'))
model.add(layers.MaxPooling2D())

model.add(Conv2D(32, (1,1), 1, activation='relu'))
model.add(layers.MaxPooling2D())

model.add(Conv2D(16, (1,1), 1, activation='relu'))
model.add(layers.MaxPooling2D())

model.add(Conv2D(32, (1,1), 1, activation='relu'))
model.add(layers.MaxPooling2D())

model.add(Conv2D(16, (1,1), 1, activation='relu'))
model.add(layers.MaxPooling2D())

model.add(Flatten())

model.add(Dense(IMG_SIZE[0], activation='relu'))
model.add(Dropout(0.5))
model.add(Dense((int(IMG_SIZE[0] / 2)), activation='relu'))
model.add(Dropout(0.5))

model.add(Dense(1, activation='sigmoid'))

model.summary()
```

```python
model.compile('adam', loss='binary_crossentropy', metrics=['accuracy'])

early_stopping = EarlyStopping(monitor="val_loss", patience=4, restore_best_weights=True)

history = model.fit(
    train_ds,
    epochs=25,
    validation_data=val_ds,
    callbacks=[early_stopping]
)
```

```python
def plot_training_history(history):
    fig, axes = plt.subplots(1, 2, figsize=(12, 5))

    axes[0].plot(history.history['loss'], label='train loss', color="black")
    axes[0].plot(history.history['val_loss'], label='val loss', color="red")
    axes[0].set_title('Loss during training')
    axes[0].set_xlabel('Epoch')
    axes[0].set_ylabel('Loss')
    axes[0].legend()

    axes[1].plot(history.history['accuracy'], label='train accuracy', color="black")
    axes[1].plot(history.history['val_accuracy'], label='val accuracy', color="red")
    axes[1].set_title('Accuracy during training')
    axes[1].set_xlabel('Epoch')
    axes[1].set_ylabel('Accuracy')
    axes[1].legend()

    plt.tight_layout()
    plt.show()

plot_training_history(history)
```

```python
def plot_images(images, labels, preds, title, correct=True, max_images=5):
    pred_classes = [int(p.item() > 0.5) for p in preds]

    filtered = [
        (img, label, pred_class)
        for img, label, pred_class in zip(images, labels, pred_classes)
        if (label == pred_class) == correct
    ]

    plt.figure(figsize=(15, 5))
    for i, (img, label, pred_class) in enumerate(filtered[:max_images]):
        plt.subplot(1, max_images, i + 1)
        plt.imshow(img.astype("int"))
        plt.title(f"Label: {label}, Pred: {pred_class}")
        plt.axis('off')

    plt.suptitle(title)
    plt.tight_layout()
    plt.show()

def extract_images_and_labels(dataset):
    images = []
    labels = []
    for img_batch, label_batch in dataset:
        images.extend(img_batch.numpy())
        labels.extend(label_batch.numpy())
    
    return np.array(images), np.array(labels)
```

```python
predictions_train = model.predict(train_ds)
model.evaluate(train_ds)

images, labels = extract_images_and_labels(train_ds)
images = images*255

plot_images(images, labels, predictions_train, "Correct Train")
plot_images(images, labels, predictions_train, "Incorrect Train", False)
```

```python
predictions_test = model.predict(test_ds)
model.evaluate(test_ds)

images, labels = extract_images_and_labels(test_ds)
images = images*255

plot_images(images, labels, predictions_test, "Correct Test")
plot_images(images, labels, predictions_test, "Incorrect Test", False)
```
